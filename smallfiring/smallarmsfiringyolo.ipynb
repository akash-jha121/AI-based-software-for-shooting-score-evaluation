```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

# Loading model
model = YOLO("/content/drive/MyDrive/Computer Vision /Small Arms firing/shootingyolo/runs/detect/train9/weights/best.pt")

# Loading image
image_path = "/content/drive/MyDrive/Computer Vision /Small Arms firing/shootingyolo/train/images/IMG1.jpg"

# Process the image through model
results = model(image_path)

for result in results:

    print(result)


image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Process the results object
# results is a list of objects; we need to access the first element to get the predictions
result = results[0]  # Access the first (and in this case, only) result
boxes = result.boxes  # Access the bounding boxes

# Iterate through the bounding boxes
for box in boxes:
    x_min, y_min, x_max, y_max = box.xyxy[0]  # Bounding box coordinates

    confidence = box.conf[0]  # Confidence score

    class_id = box.cls[0]  # Class ID

    # Convert coordinates to integers
    x_min, y_min, x_max, y_max = int(x_min), int(y_min), int(x_max), int(y_max)

    # Draw the bounding box
    cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (255, 0, 0), 2)

    # Optionally, add the confidence score and class label
    label = f"Class: {int(class_id)}, Confidence: {confidence:.2f}"
    cv2.putText(image, label, (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

# Display the image
plt.figure(figsize=(10, 30))
plt.imshow(image)
plt.axis("off")
plt.show()


```
